```python
pd.read_sql("""

WITH popular_movies
AS
(
SELECT 
       b.primary_title, r.numvotes, r.averagerating, b.genres
FROM 
       movie_basics as b
INNER JOIN 
       movie_ratings as r
ON b.movie_id = r.movie_id

ORDER BY r.numvotes DESC

LIMIT 500
)

SELECT primary_title as 'Title', averagerating as 'Average Rating', numvotes as 'Num. of Votes', genres as 'Genres'
FROM
    popular_movies
ORDER BY averagerating DESC;

""", conn)
```

```python
pd.read_sql("""
SELECT 
      b.movie_id, d.person_id, b.primary_title, r.numvotes, r.averagerating, b.genres
FROM 
       movie_basics as b
INNER JOIN 
       movie_ratings as r
ON 
       b.movie_id = r.movie_id
LEFT JOIN 
       directors as d
ON
       r.movie_id = d.movie_id


ORDER BY r.numvotes DESC

LIMIT 500

""", conn)
```
